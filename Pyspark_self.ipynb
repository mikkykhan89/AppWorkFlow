```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
import os

# Initialize SparkSession
spark = SparkSession.builder.appName('pyspark_self').getOrCreate()

# schema
schema = StructType([
    StructField("EmpName", StringType(), True),
    StructField("Dept", StringType() , True),
    StructField("Salary", DoubleType() , True),
    ])

ddl_schema = "EmpName String, Dept String, Salary Double"

# Data
Data = [
    ("DATA-ABC3", 'Data Engineer', 300000.0),
    ("DATA-ABC2", 'Data Engineer', 200000.0),
    ("DATA-ABC1", 'Data Engineer', 100000.0),
    ("SUPP-XYZ3", 'Support Engineer', 100000.0),
    ("SUPP-XYZ2", 'Support Engineer', 90000.0),
    ("SUPP-XYZ1", 'Support Engineer', 80000.0),
    ("TEST-MNL3", 'Tester Engineer', 100000.0),
    ("TEST-MNL2", 'Tester Engineer', 90000.0),
    ("TEST-MNL1", 'Tester Engineer', 80000.0),
    ]

#print(Data)

# Fixed typo: createDataFrame (capital F)
df = spark.createDataFrame(Data, schema=ddl_schema)
#df.show()

#df.filter(col("Salary")>200000).show()

#df.createOrReplaceTempView("tblsalary")
#spark.sql("select * from (select *, row_number() over(partition by dept order by dept,salary desc) as rn from tblsalary) a where rn <=2").show()

df_newcol = df.withColumn("sub_dept",lit("Tech"))

df_newcol.show()

```

### Common Syntax to Add Columns in PySpark

Here are the standard ways to add columns to a DataFrame:

```python
from pyspark.sql.functions import col, lit, when

# 1. Add a constant value
df_example = df.withColumn("Status", lit("Active"))

# 2. Add a column based on a calculation of existing columns
df_example = df_example.withColumn("Bonus", col("Salary") * 0.1)

# 3. Add a column using conditional logic (Case When)
df_example = df_example.withColumn("Seniority",
    when(col("Salary") >= 200000, "High")
    .when(col("Salary") >= 100000, "Medium")
    .otherwise("Low")
)

# 4. Chain multiple additions
df_example = df_example.withColumn("Country", lit("USA")) \
                       .withColumn("Currency", lit("USD"))

df_example.show()
```
